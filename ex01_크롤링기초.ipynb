### BeautifulSoup ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
- html ë¬¸ì„œì—ì„œ ì›í•˜ëŠ” ë°ì´í„°ë¥¼ ì¶”ì¶œí•˜ê¸° ì‰½ê²Œ í•´ì£¼ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬

```python
from bs4 import BeautifulSoup as bs 
# ì»´í“¨í„°ê°€ ì´í•´í•  ìˆ˜ ìˆëŠ” íŒŒì´ì¬ ê°ì²´ í˜•íƒœë¡œ ë³€í™˜í•´ì£¼ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬
```

```python
res = req.get('https://www.kakaocorp.com/page/service/service/KakaoTalk')
res
```

```python
res.text   # html ë¬¸ì„œ ë¶ˆëŸ¬ì˜¤ë©´ ë³´ê¸°ê°€ ì–´ë ¤ì›€ --> Tag Soup ìƒíƒœ
           # ë³´ê¸° í˜ë“œë‹ˆê¹Œ ì˜ˆì˜ê²Œ ì •ë¦¬í• ë˜ --> Beautiful Soup
```

```python
# bs(parsing(í•´ì„) í•  html, 'parsingí•  ë°©ë²•')
# íŒŒì´ì¬ì—ì„œ ì œê³µí•˜ëŠ” parsing ë°©ë²•ì€ ì´ 4ê°€ì§€
# lxml : ë¹ ë¥´ê³  ê´€ëŒ€í•˜ê¸° ë•Œë¬¸ì—
soup = bs(res.text, 'lxml')   # íŒŒì´ì¬ ê°ì²´ í˜•íƒœë¡œ ë³€í™˜! (íŒŒì´ì¬ì´ ì½ì–´ë“¤ì´ê¸° ì‰½ê²Œ ë³€í™˜)
soup
```

#### ì¹´ì¹´ì˜¤í†¡ í˜ì´ì§€ ì •ë³´ë¥¼ ê°€ì ¸ì™€ë³´ì
 - ì¹´ì¹´ì˜¤í†¡ í˜ì´ì§€ì—ì„œ ê¸€ìë¥¼ ìˆ˜ì§‘í•´ë³´ì

- ìš”ì†Œ(Element) = íƒœê·¸(tag) + ë‚´ìš©(contents)
    - íƒœê·¸ : ìœ„ì¹˜!
    - ë‚´ìš© : ìˆ˜ì§‘í•  contents

```python
# 'ì¹´í†¡ ì•ˆë…•ê°€ì´ë“œ' ë¼ëŠ” ë¬¸ìë¥¼ ìˆ˜ì§‘
# íƒœê·¸ì˜ ìœ„ì¹˜ë¥¼ ì°¾ëŠ” ë°©ë²•
# 1. ë‚´ê°€ ê°€ì ¸ì˜¤ê³  ì‹¶ì€ contents ìœ„ì— ë§ˆìš°ìŠ¤ë¥¼ ì˜¬ë ¤ë‘ê³  ìš°í´ë¦­ -> ê²€ì‚¬
# 2. ê°œë°œìëª¨ë“œ ìƒë‹¨ì— ì»¤ì„œ ëª¨ì–‘ ì•„ì´ì½˜ í´ë¦­ í›„ ê°€ì ¸ì˜¤ê³  ì‹¶ì€ ì½˜í…ì¸  í´ë¦­
```

```python
# ì¹´í†¡ ì•ˆë…•ê°€ì´ë“œ ë¼ëŠ” ê¸€ìëŠ” ì–´ë–¤ íƒœê·¸ ì•ˆì—?
# aíƒœê·¸ ê°€ì ¸ì˜¤ëŠ” ë°©ë²•
soup.select('a')
# í˜ì´ì§€ì— ìˆëŠ” ëª¨ë“  aíƒœê·¸ê°€ ì „ë¶€ ê°€ì ¸ì™€ì¡Œë‹¤.
```

```python
# aíƒœê·¸ ì¤‘ì—ì„œ ìš°ë¦¬ê°€ ì›í•˜ëŠ” íƒœê·¸ë§Œ ê°€ì ¸ì˜¤ë ¤ë©´ ì„ íƒì(êµ¬ë¶„ì) í•„ìš”!
# aíƒœê·¸ ì¤‘ì—ì„œ í´ë˜ìŠ¤ê°€ link_directì¸ ì• ë“¤ë§Œ ë°ë ¤ì˜¬ê±°ì•¼!
# í´ë˜ìŠ¤ì— ì ‘ê·¼í•˜ëŠ” ê¸°í˜¸ëŠ” ì˜¨ì (.)ì„ ì‚¬ìš©
soup.select('a.link_direct')  # --> ë°ì´í„° í˜•ì‹ : list

# select : ì—¬ëŸ¬ê°€ì§€ ìš”ì†Œë¥¼ ë³µìˆ˜í˜•ìœ¼ë¡œ ê°€ì ¸ì™€ ì¤€ë‹¤. (í•­ìƒ list í˜•íƒœë¡œ)
```

```python
# select_one : ì—¬ëŸ¬ê°€ì§€ ìš”ì†Œ ì¤‘ì—ì„œ ë§¨ ìœ„ì— ìˆëŠ” ì²«ë²ˆì§¸ ìš”ì†Œë§Œ ë‹¨ìˆ˜í˜•ìœ¼ë¡œ ê°€ì ¸ì™€ ì¤€ë‹¤. 
soup.select_one('a.link_direct')
```

```python
# ì¹´í†¡ ì•ˆë…•ê°€ì´ë“œ ê¸€ìë§Œ ìˆ˜ì§‘
a = soup.select('a.link_direct')[0]
print(a.text)     # .text : ìš”ì†Œ ë‚´ë¶€ì— ìˆëŠ” contentsë§Œ ê°€ì ¸ì˜¤ëŠ” ê¸°ëŠ¥
# ë°ì´í„° ì• ë’¤ì— ê³µë°±ì´ ë§ë‹¤?
```

```python
print(a.text.strip())
```
