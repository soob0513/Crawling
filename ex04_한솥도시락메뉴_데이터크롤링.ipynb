````python
# 필요한 라이브러리 로딩
from selenium import webdriver as wb
# 웹 브라우저를 제어하기 위한 라이브러리
from selenium.webdriver.common.keys import Keys
# 웹에게 값을 '입력'받기 위한 라이브러리>> 컴퓨터용 키보드
from selenium.webdriver.common.by import By
# html 문서에서 태그와 선택자의 위치를 찾기 위한 라이브러리
from bs4 import BeautifulSoup as bs
# 컴퓨터가 이해하기 쉬운 객체로 변환해주는 라이브러리
````

````python
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')
````

````python
# bs 객체화 : bs(parsing할 html 문서, 'parsing 방법')
# 동적 페이지 크롤링 -> res.text가 존재하지 않는다!
# html 문서에 대한 정보 가져오기
driver.page_source   # 문자열 형태의 html 문서가 가져와졌다!
````

#### 예외 처리
- try : 오류 나기 전의 로직 실행
- except : 오류가 났을 때 로직 실행

````python
import time   # 시간 관련 라이브러리(딜레이하는 용도!)
````

````python
while True : 
    try :  # 일단 시도해봐! 

        # 더보기 버튼 클릭
        btn = driver.find_element(By.CSS_SELECTOR,'#btn_more > span > a')
        btn.click()

        time.sleep(1)     # 1초 딜레이!
        print("더보기 클릭!")
    except :   # 오류가 난 경우! -> 더보기 버튼이 다 클릭된 경우
        print("더보기 완료!")
        break
````

#### 실습! 한 셀에 완벽한 크롤링 코드 짜기!
1. 한솥메뉴 사이트 요청
2. 더보기 버튼 끝까지 누르기
3. 가격과 메뉴명 요소 수집
4. content 추출
5. DataFrame 제작 & csv 형식 저장
6. chrome 닫기 (자원 반납)
7. 한 셀로 합치기! (Shift + m) Shift 누른 상태로 방향키누르기+m
    - ※ 셀 나누기 Ctrl + Shift + -

````python
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')

# bs 객체화 (parsing)
soup = bs(driver.page_source, 'lxml')
soup

#btn_more > span > a
btn = driver.find_element(By.CSS_SELECTOR, '#btn_more > span > a')
btn

while True : 
    try : 
        btn = driver.find_element(By.CSS_SELECTOR, '#btn_more > span > a')
        btn.click()

        time.sleep(1)
        print('더보기 클릭!')
    except : 
        print('더보기 완료!')
        break

#menuList_544 > div > div.item-text > h4
menu = driver.find_elements(By.CSS_SELECTOR, 'h4.h.fz_03')

#menuList_544 > div > div.item-text > div > strong
price = driver.find_elements(By.CSS_SELECTOR, 'div.item-text > div > strong')

menu_lst = []
price_lst = []
ind = []
for i in range(len(menu)) : 
    menu_lst.append(menu[i].text)
    price_lst.append(price[i].text)
    ind.append(i+1)

dic = {'메뉴' : menu_lst, '가격' : price_lst, ' ' : ind}

import pandas as pd
hsd = pd.DataFrame(dic)
hsd.set_index(' ', inplace=True)

hsd.to_csv('한솥도시락메뉴.csv', encoding = 'utf-8')

driver.quit()
````

````python
# 선생님 풀이
# 1. 한솥메뉴 사이트 요청
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')
print('1단계 완료')

# 2. 더보기 버튼 끝까지 누르기
while True : 
    try : 
        btn = driver.find_element(By.CSS_SELECTOR, '#btn_more > span > a')
        btn.click()

        time.sleep(1)
        print('더보기 클릭!')
    except : 
        print('더보기 완료!')
        break

        
# 3. 가격과 메뉴명 요소 수집
menu = driver.find_elements(By.CSS_SELECTOR, 'h4.h.fz_03')
price = driver.find_elements(By.CSS_SELECTOR, 'div.item-text > div > strong')
print('3단계 완료')

# 4. content 추출
# append()는 정말 많이 쓰임!
# 리스트 컴프리헨션(리스트 내포, List Comprehension)

menu_list = [menu[i].text for i in range(len(menu))]
price_list = [price[i].text for i in range(len(price))]
print('4단계 완료')

# 5. DataFrame 제작 & csv 형식 저장\
import pandas as pd

dic = {'메뉴' : menu_list, '가격' : price_list}

hansot = pd.DataFrame(dic)
hansot

hansot.to_csv('한솥메뉴.csv', encoding ='euc-kr')
print('5단계 완료')

# 6. chrome 닫기 (자원 반납)
driver.quit()
print('6단계 완료')

# 7. 한 셀로 합치기
````
